Esto se puede reescribir de forma compacta utilizando **Ax=b**, donde **A** es una *matriz de dimensiones nxn* y **x** y **b** son *vectores columna de longitud n*.

# Ejemplo:
-Defina el siguuiente sistema de ecuaciones



```python
import numpy as np
A = np.matrix([[2,3], [1,-2]])
b = np.matrix([[8],[-10]])
print("Esto es A \n",A)
print("Esto es b \n",b)
```

* Empleando la ecuación **Ax=b**, se despeja **x** y se asume que la matriz **A** es regular, se puede multiplicar la expresión por la inversa de **A**,**A-1**



* Por tanto **A-1A** es la matriz identidad y se puede escribir lla solución del sistema como:



```python
x = (A**-1)*b

print("Inverso A-1\n", x)
```

Determinante

```python
print(np.linalg.det(A))
```

Recordemos...
Un **determinante es un número que puede calcularse sobre llas matrices cuadradas.** Su valor se da a partir de la suma de los  pproductos de las diaglonales de la matriz en dirección menos la suma de los productos de las diagonales en la otra dirección y se representa con el simbolo **|A|.**




- Además una matriz se considera singular solo si su determinante es nulo y, en ese caso, no existe su matriz inversa.





Empleando



- Identifique la forma matricial empleamdo **Ax=b**
- Calcule el valor de **x**
- Calcule su determinante 

```python
A4 = np.matrix([[3, 2, -1], [2, -2, 4], [-1, 1/2, -1]])
b4 = np.matrix([[1], [-2], [0]])

print("Este es A\n", A4)
print("Este es b\n", b4)
```

```python
x4 = (A4**-1)*b4
print(x4)
```

Determinante:

```python
print(np.linalg.det(A4))
```

# Actividad:


```python
import numpy as np

# Matrices y cálculos del determinante
A = np.matrix([[9]])
B = np.matrix([[4, -1], [-2, 0]])
C = np.matrix([[5, 0, 2], [3, 1, 1], [0, 1, 2]])


print("\nEste es A\n", A)
print("\nEste es B\n", B)
print("\nEste es C\n", C)
```

Determinantes

```python
# Determinantes
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_C = np.linalg.det(C)

print("Determinante de A:", det_A)
print("Determinante de B:", det_B)
print("Determinante de C:", det_C)
```

# Nucleo de una matriz:

```python
import numpy as np
from scipy.linalg import null_space

# Definimos la matriz A
A = np.array([[1, 1, 0],
              [1, 1, 0]])

# Calculamos el kernel (núcleo) de la matriz A
kernel_A = null_space(A)
kernel_A
```

```python
A2 = np.array([[1, 1, 2],
                  [2, 2, 4],
                  [2, 3, 5]])

# Calculamos el kernel (núcleo) de la nueva matriz A
kernel_A2 = null_space(A2)
kernel_A2
```
