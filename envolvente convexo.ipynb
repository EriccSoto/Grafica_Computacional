# Funciones

### Explicación ampliada del código de funciones

1. **Función `turn_right()`**:
   - Esta función se encarga de determinar si el giro entre tres puntos consecutivos es en sentido antihorario o en sentido horario utilizando el determinante. Si es antihorario, el punto intermedio se conserva, pero si es horario, se elimina, asegurando que la envolvente convexa siga una forma "convexa", eliminando los puntos que no son parte del contorno exterior.

2. **Función `convex_hull()`**:
   - Primero, se ordenan los puntos en función de su coordenada x. Luego, se dividen en dos partes: la parte superior y la inferior del contorno. La función invierte la lista para procesar la parte inferior, aplicando el mismo proceso de "turn_right" en ambas partes y luego combinando los resultados para formar la envolvente convexa completa.

3. **Función `graph()`**:
   - La función `graph()` se encarga de crear la visualización utilizando la biblioteca `matplotlib`. Extrae las coordenadas de los puntos y de la envolvente convexa, ajusta los límites de la gráfica para que todos los puntos se muestren correctamente, y finalmente dibuja los puntos como círculos negros y la envolvente convexa como una línea verde que conecta los puntos del contorno.

```python
def turn_right():
    array =[coord_points[0], coord_points[1]]
    for i in range(2, len(coord_points)):
        array.append(coord_points[i])
        while len(array) > 2 and np.linalg.det([array[-3], array[-2],array[-1]]) > 0:
            array.pop(-2)
    return array
    
def convex_hull():
    coord_points.sort()
    l_upper = turn_right()
    coord_points.reverse()
    l_lower = turn_right()
    l = l_upper + l_lower
    return l


def graph(convex_pol, coord_points):
    
    x_points = [i[0] for i in coord_points]
    y_points = [i[1] for i in coord_points]

    x_polygon = [i[0] for i in convex_pol]
    y_polygon = [i[1] for i in convex_pol]

    x_lim_der = max(x_points) + 5
    y_lim_sup = max(y_points) + 5

    x_lim_izq = min(x_points) - 5
    y_lim_inf = min(y_points) - 5

    plt.xlim(x_lim_izq, x_lim_der)
    plt.ylim(y_lim_inf, y_lim_sup)


    plt.title('Problema: convex Hull')
    plt.xlabel('Eje de las abcisas')
    plt.ylabel('Eje de las ordenadas')
    plt.plot(x_points, y_points, 'ko')
    plt.plot(x_polygon, y_polygon, 'g-', linewidth = 3.0)
    plt.show()
```

# Octágono
El código genera un octágono regular a partir de puntos específicos y los procesa para obtener su envolvente convexa. La función `turn_right()` asegura que los puntos se organizan en sentido antihorario al calcular el determinante de matrices 3x2. `convex_hull()` combina los puntos superiores e inferiores para crear la envolvente convexa. Finalmente, la función `graph()` grafica los puntos originales y la envolvente, añadiendo etiquetas para mejorar la visualización y ajustando los límites de la gráfica según los valores máximos y mínimos de los puntos.

```python
import random as rand
import numpy as np
import matplotlib.pyplot as plt

def turn_right():
    array = [coord_points[0], coord_points[1]]
    for i in range(2, len(coord_points)):
        array.append(coord_points[i])
        while len(array) > 2:
            v1 = np.array(array[-3])
            v2 = np.array(array[-2])
            v3 = np.array(array[-1])
            det = np.linalg.det([v1 - v2, v3 - v2])
            if det > 0:
                array.pop(-2)
            else:
                break
    return array

def convex_hull():
    coord_points.sort()
    l_upper = turn_right()
    coord_points.reverse()
    l_lower = turn_right()
    l = l_upper + l_lower
    return l

def graph(convex_pol, coord_points):
    x_points = [i[0] for i in coord_points]
    y_points = [i[1] for i in coord_points]

    x_polygon = [i[0] for i in convex_pol]
    y_polygon = [i[1] for i in convex_pol]

    x_lim_der = max(x_points) + 5
    y_lim_sup = max(y_points) + 5
    x_lim_izq = min(x_points) - 5
    y_lim_inf = min(y_points) - 5

    plt.xlim(x_lim_izq, x_lim_der)
    plt.ylim(y_lim_inf, y_lim_sup)

    plt.title('Octágono: Convex Hull')
    plt.xlabel('Eje de las abscisas')
    plt.ylabel('Eje de las ordenadas')
    plt.plot(x_points, y_points, 'ko', label="Puntos del Octágono")
    plt.plot(x_polygon, y_polygon, 'g-', linewidth=3.0, label="Envolvente Convexa")
    plt.legend()

coord_points = [
    [30, 100],
    [50, 80],
    [60, 60],
    [50, 40],
    [30, 20],
    [10, 40],
    [0, 60],
    [10, 80]
]

convex_pol = convex_hull()
graph(convex_pol, coord_points)

plt.legend()
plt.show()
```
