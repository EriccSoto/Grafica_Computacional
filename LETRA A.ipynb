```python
import numpy as np
import matplotlib.pyplot as plt


A = np.array([
    (0, 0, 1),(0.5,0,1),(1,1.5,1),(2,1.5,1),(2.5,0,1),(3,0,1),(1.75,4,1),(1.25,4,1),(0,0,1),
])

A2 =  np.array([(1.1,2,1),(1.5,3,1),(1.9,2,1),(1.1,2,1)])
```

```python
# Matriz Identidad
Ic = np.eye(3)

# Graficamos la letra "A" original
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(A[:, 0], A[:, 1], color="blue", label="Letra A Original")
ax.plot(A2[:, 0], A2[:,1])
ax.set_title("Letra E Original")
ax.set_xticks(np.arange(-1, 5, 1))
ax.set_yticks(np.arange(-1, 6, 1))
ax.set_aspect('equal', adjustable='box')
ax.grid()
ax.legend()
plt.show()
```

```python
#REFLEXIÓN EN EJE X

Refx = np.array([
    [1., 0, 0],
    [0, -1., 0],
    [0., 0., 1.]
])

# Aplicamos la reflexión con respecto al eje x
Ax, Ay = [], []
A2x,A2y = [],[]

for row in A:
    output_row = Refx.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    Ax.append(x)
    Ay.append(y)

for row in A2:
    output_row = Refx.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    A2x.append(x)
    A2y.append(y)


# Graficamos la transformación
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(A[:, 0], A[:, 1], color="blue", label="Letra A Original")
ax.plot(A2[:, 0], A2[:,1])
ax.plot(Ax, Ay, color="red", label="Reflexión respecto al eje x")
ax.plot(A2x,A2y,color="red")
ax.set_title("Reflexión respecto al eje X")
ax.set_xticks(np.arange(-1, 6, 1))
ax.set_yticks(np.arange(-6, 6, 1))
ax.set_aspect('equal', adjustable='box')
ax.grid()
ax.legend()
plt.show()
```

```python
# Matriz de Reflexión con respecto al eje y
Refy = np.array([[-1., 0, 0], [0, 1., 0], [0., 0., 1.]])

# Aplicamos la reflexión con respecto al eje y
Ax, Ay = [], []
A2x,A2y = [],[]

for row in A:
    output_row = Refy.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    Ax.append(x)
    Ay.append(y)

for row in A2:
    output_row = Refy.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    A2x.append(x)
    A2y.append(y)


# Graficamos la transformación
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(A[:, 0], A[:, 1], color="blue", label="Letra A Original")
ax.plot(A2[:, 0], A2[:,1])
ax.plot(Ax, Ay, color="red", label="Reflexión respecto al eje x")
ax.plot(A2x,A2y,color="red")
ax.set_title("Reflexión respecto al eje X")
ax.set_xticks(np.arange(-4, 6, 1))
ax.set_yticks(np.arange(-1, 6, 1))
ax.set_aspect('equal', adjustable='box')
ax.grid()
ax.legend()
plt.show()
```

```python
# Matriz de deformación horizontal/vertical
h = -1
v = 0
D = np.array([
    [1., h, 0.],
    [v, 1., 0.],
    [0., 0., 1.]
])

Ax, Ay = [], []
A2x,A2y = [],[]

for row in A:
    output_row = D.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    Ax.append(x)
    Ay.append(y)

for row in A2:
    output_row = D.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    A2x.append(x)
    A2y.append(y)


# Graficamos la transformación
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(A[:, 0], A[:, 1], color="blue", label="Letra A Original")
ax.plot(A2[:, 0], A2[:,1])
ax.plot(Ax, Ay, color="red", label="Reflexión respecto al eje x")
ax.plot(A2x,A2y,color="red")
ax.set_title("Reflexión respecto al eje X")
ax.set_xticks(np.arange(-4, 6, 1))
ax.set_yticks(np.arange(-1, 7, 1))
ax.set_aspect('equal', adjustable='box')
ax.grid()
ax.legend()
plt.show()
```

```python
# Matriz de traslación
tx = -3
ty = -5
T = np.array([
    [1., 0, tx],
    [0, 1., ty],
    [0., 0., 1.]
])

# Aplicamos la traslación
Ax, Ay = [], []
A2x,A2y = [],[]

for row in A:
    output_row = T.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    Ax.append(x)
    Ay.append(y)

for row in A2:
    output_row = T.dot(row)  # Multiplicación matriz-vector
    x, y = output_row[:2]    # Extraemos las coordenadas x, y
    A2x.append(x)
    A2y.append(y)
    


# Graficamos la transformación
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(A[:, 0], A[:, 1], color="blue", label="Letra A Original")
ax.plot(A2[:, 0], A2[:,1])
ax.plot(Ax, Ay, color="red", label="Reflexión respecto al eje x")
ax.plot(A2x,A2y,color="red")
ax.set_title("Reflexión respecto al eje X")
ax.set_xticks(np.arange(-4, 6, 1))
ax.set_yticks(np.arange(-6, 6, 1))
ax.set_aspect('equal', adjustable='box')
ax.grid()
ax.legend()
plt.show()
```
