# Importación de librerías necesarias

En esta celda, se importan las librerías `numpy` y `matplotlib.pyplot`. 
Estas se usarán para manejar matrices y graficar las transformaciones de la letra "E".


```python
import numpy as np
import matplotlib.pyplot as plt
```

# Definición de la matriz "E"

Esta celda define la matriz que representa la letra "E" en coordenadas homogéneas. 
La matriz contiene los puntos de la "E", incluidos los necesarios para cerrar la figura y formar la línea faltante.


```python
E = np.array([
    (0, 0, 1), (3, 0, 1), (3, 1, 1), (1, 1, 1), 
    (1, 2, 1), (2, 2, 1), (2, 3, 1), (1, 3, 1), 
    (1, 4, 1), (3, 4, 1), (3, 5, 1), (0, 5, 1), 
    (0, 0, 1)  # Línea de cierre superior a inferior
])
```

# Definición de matrices de transformación

```python
# Matriz Identidad
Ic = np.eye(3)

# Matriz de Reflexión con respecto al eje x
Refx = np.array([[1., 0, 0], [0, -1., 0], [0., 0., 1.]])

# Matriz de Reflexión con respecto al eje y
Refy = np.array([[-1., 0, 0], [0, 1., 0], [0., 0., 1.]])

# Matriz de Rotación
theta = np.pi / 3  # Ángulo de rotación deseado
R = np.array([
    [np.cos(theta), np.sin(theta), 0.],
    [-np.sin(theta), np.cos(theta), 0.],
    [0., 0., 1.]
])

# Matriz de cambio de escala
s = 2  # Escalar
S = np.array([
    [s, 0, 0.],
    [0., s, 0.],
    [0., 0., 1.]
])

# Matriz de deformación horizontal/vertical
h = -1
v = 0
D = np.array([
    [1., h, 0.],
    [v, 1., 0.],
    [0., 0., 1.]
])

# Matriz de traslación
tx = -3
ty = -5
T = np.array([
    [1., 0, tx],
    [0, 1., ty],
    [0., 0., 1.]
])
```

# Función para aplicar y graficar transformaciones

En esta celda, se define una función que aplica una transformación a la matriz de puntos de la "E" y 
grafica tanto la figura original como la transformada en una cuadrícula.


```python
def graficar_transformacion(E, transformacion, titulo):
    Ex, Ey = [], []
    for row in E:
        output_row = transformacion.dot(row)  # Aplicar transformación
        x, y, _ = output_row
        Ex.append(x)
        Ey.append(y)
    
    # Graficar letra original y transformada
    plt.figure(figsize=(8, 8))
    plt.plot(E[:, 0], E[:, 1], color="blue", label="Original")
    plt.plot(Ex, Ey, color="red", label=titulo)
    ax = plt.gca()
    ax.set_xticks(np.arange(-6, 8, 2))
    ax.set_yticks(np.arange(-6, 8, 2))
    ax.set_aspect('equal', adjustable='box')
    plt.title(f"Transformación: {titulo}")
    plt.grid()
    plt.legend()
    plt.show()
```

# Aplicación y graficación de transformaciones
En esta celda, se llama a la función `graficar_transformacion` para aplicar y graficar cada transformación. 
Cada transformación se aplica a la matriz "E" y se visualiza en una gráfica individual.


```python
# Reflexión en el eje X
graficar_transformacion(E, Refx, "Reflexión en X")


```

```python
# Reflexión en el eje Y
graficar_transformacion(E, Refy, "Reflexión en Y")
```

```python

# Rotación de π/3
graficar_transformacion(E, R, "Rotación π/3")

```

```python
# Escalado por un factor de 2
graficar_transformacion(E, S, "Escalado por 2")
```

```python

# Deformación horizontal/vertical
graficar_transformacion(E, D, "Deformación (h=-1, v=2)")
```

```python

# Traslación (-3, -5)
graficar_transformacion(E, T, "Traslación (-3, -5)")
```
