### Escalado
se grafica el triangulo en su forma original así como el cuadrado original, para luego aplicar un aumento de su tamaño 5 veces

```python
cuadro_original=([1,1,2,2,1],[1,2,2,1,1])
triangulo_original=([1,1.5,2,1],[1,2,1,1])
cuadro_escalado=([1*5,1*5,2*5,2*5,1*5],[1*5,2*5,2*5,1*5,1*5])
triangulo_escalado=([1*5,1.5*5,2*5,1*5],[1*5,2*5,1*5,1*5])
plt.xlim(0,12)
plt.ylim(0,12)
plt.plot(cuadro_original[0],cuadro_original[1])
plt.plot(cuadro_escalado[0],cuadro_escalado[1])

plt.plot(triangulo_original[0],triangulo_original[1])
plt.plot(triangulo_escalado[0],triangulo_escalado[1])
```

### Traslación

Se grafican los elementos en un punto inicial y su nuevo ppunto de traslación


```python
cuadro_original=([1,1,2,2,1],[1,2,2,1,1])
triangulo_original=([1,1.5,2,1],[1,2,1,1])
cuadro_trasladado=([1+4,1+4,2+4,2+4,1+4],[1+4,2+4,2+4,1+4,1+4])
triangulo_trasladado=([1+4,1.5+4,2+4,1+4],[1+4,2+4,1+4,1+4])
plt.xlim(0,7)
plt.ylim(0,7)
plt.plot(cuadro_original[0],cuadro_original[1])
plt.plot(cuadro_trasladado[0],cuadro_trasladado[1])

plt.plot(triangulo_original[0],triangulo_original[1])
plt.plot(triangulo_trasladado[0],triangulo_trasladado[1])
```

### rotación

Se grafican los elementos en unanposición inicial y se grafica su rotación


```python
teta=np.radians(45)
cuadro_originalX=np.array([1,1,2,2,1])
cuadro_originalY=np.array([1,2,2,1,1])
triangulo_originalX=np.array([1,1.5,2,1])
triangulo_originalY=np.array([1,2,1,1])

tx_rotado=np.cos(teta) * triangulo_originalX - np.sin(teta) * triangulo_originalY
ty_rotado=np.sin(teta) * triangulo_originalX + np.cos(teta) * triangulo_originalY

cx_rotado=np.cos(teta) * cuadro_originalX - np.sin(teta) * cuadro_originalY
cy_rotado=np.sin(teta) * cuadro_originalX + np.cos(teta) * cuadro_originalY




plt.plot(tx_rotado,ty_rotado)
plt.plot(cx_rotado,cy_rotado)

plt.plot(cuadro_originalX,cuadro_originalY)
plt.plot(triangulo_originalX,triangulo_originalY)
```
